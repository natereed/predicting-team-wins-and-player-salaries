```python
# Get rid of entries for multiple teams, replace with the aggregated (roll-up) entry provided
# with team = "n teams".

batting_df = pd.read_csv(os.path.join("..", "data", "db", "batting.csv"))
batting_df['Year'] = batting_df.Year.str.extract("^([0-9]{4})")

# Find the "2 team" entries, replace the year with a valid numeric entry, and delete the other
# individual team entries
#multiple_teams_df = batting_df[batting_df.Year.str.contains("[-]")] 
multiple_teams_df = batting_df[batting_df.Team.str.contains("teams")]
for index, row in multiple_teams_df.iterrows():
    print("Year: " + row['Year'])
    print("Player: " + row['Player Id'])
    print("Team: " + row['Team'])
    # row['Year] and row['Player Id'] are the ones to keep
    
    batting_df = batting_df[((batting_df['Year'] != row['Year']) | (batting_df['Player Id'] != row['Player Id']))
                            | ((batting_df['Year'] == row['Year']) & (batting_df['Player Id'] == row['Player Id']) 
                            & (batting_df['Team'] == row['Team']))]
    
batting_df['Year'] = pd.to_numeric(batting_df['Year'])
```

```python
salaries_df = salaries_df[salaries_df['Year'] >= 2015].head(10)

grouped_salaries = salaries_df.groupby(["Player Id", "Year"])
grouped_salaries
```

```python
# Generate dict with player-year data
# Each player is an entry with keys: "Player Id", "Year", "Year.G", "Year.AB",.... "Year-1.G", "Year-1.AB" etc

stats = {}
for name, group in grouped_salaries:
    player_id = name[0] # short name, like 'bcolon'
    salary_year = name[1]
    stats[player_id] = {'salary_year' : salary_year}
    print("Player {} in salary year {}".format(player_id, salary_year))
    player_df = batting_df[(batting_df['Player Id'] == player_id) & (batting_df['Year'] <= salary_year)]
    
    # Iterate over performance stats for the given player
    # Each row is for a different year. Assemble all years into a single row.
    # Player Id, Year, Team, LG, Year.G, Year.AB,... Year-1.G, Year-1.AB, etc.
    for index, year_row in player_df.iterrows():
        play_year = year_row['Year']
        print("Adding data for {} in year {}".format(player_id, play_year))
        for column in batting_df.columns[4:].values:
            year_diff = salary_year - play_year
            if (year_diff == 0):
                years_relative = ""
            else:
                years_relative = "-{}".format(year_diff)
            stat_name = "Year{}_{}".format(years_relative, column)    
            stats[player_id][stat_name] = year_row[column]
            
# Create data frame from dictionary

# Merge with Salaries dataframe on player_id and year
stats        
```

```python
obs = []
for index, row in batting_df.iterrows():
    observations = {}
    year=row['Year']
    
    # Create series
    for column in batting_df.columns[:4].values:
        observations[column] = pd.Series()
    for column in batting_df.columns[4:].values:
        observations[str(year) + "." + column] = pd.Series()
        
    # Copy values of row to series
    for column in batting_df.columns[:4].values:
        series = observations[column] 
        observations[column] = series.append(pd.Series(row[column]))      
            
    for column in batting_df.columns[4:].values:
        column = str(year) + "." + column
        series = observations[column]
        observations[column] = series.append(pd.Series(row[column]))      

    obs.append(observations)

obs
```

```python
obs_df = pd.DataFrame(observations)
```
